```python
import pandas as pd
import numpy as np
```

```python
flights_url = "https://raw.githubusercontent.com/roberthryniewicz/datasets/master/airline-dataset/flights/flights.csv"
flights=pd.read_csv(flights_url)
print(flights)
```

# Missing_data

```python
flights.isnull().sum(axis=0)
```

```python
flights[flights['DepTime'].isnull()]
```

```python
flights[flights['TaxiOut'].isnull()]
```

# Sorting_data

```python
flights['Dest'].sort_values()
```

```python
flights['Origin'].sort_values(ascending=False)
```

```python
flights.sort_values(['Origin','FlightNum'])
```

# Grouping_data

```python
flights['Distance'].mean()
```

```python
flights.groupby('Origin').size()
```

```python
flights.groupby('Origin').Distance.mean()
```

```python
flights[flights.Dest=='EWR'].Distance.max()
```

```python
flights.groupby('Dest').Distance.agg(['count','min','max','mean'])
```

```python
%matplotlib inline
flights.groupby('Year').mean().plot(kind='bar')
```
